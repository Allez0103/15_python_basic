```python
## 값을 여러개 출력하기

print(1, 2, 3)
```

```python
## sep로 값 사이에 문자넣기

print(1, 2, 3, sep=", ")
```

```python
print(1920, 1080, sep=" X ")
```

```python
## 줄바꿈 활용하기

print(1, 2, 3, sep="\n")
```

```python
print("1\n2\n3")
```

```python
## end 사용하기

print(1)
print(2)
print(3)
```

```python
print(1, end=", ")
print(2, end=", ")
print(3, end="")
print("n")
```

```python
print(1, end=" ")
print(2, end=" ")
print(3)
```
