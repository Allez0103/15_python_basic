```python
## 딕셔너리

# 딕셔너리는 키와 밸류로 구성

lux = {
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}

print(type(lux))
print(lux)
```

```python
# 키 이름이 중복이 되면

lux = {
    "health" : 490,
    "health" : 800,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}

print(lux)
print(lux["health"])
# 키 값이 중복될경우 뒤의 값
```

```python
# 딕셔너리 키의 자료형

# 1. 문자열
# 2. 정수
# 3. 실수
# 4. bool (True, False)
# 5. 튜플
# immutable 한 자료형

# 키로 사용할 수 없는 자료형

# 1. 딕셔너리
# 2. 리스트
```

```python
x = {
    100 : "hundred",
    False : 0,
    3.5 : [3.5, 3.5]
}

print(type(x))
print(x)
```

```python
# 빈 딕셔너리 만들어서 사용

x = {}
y = dict()

print(x)
print(type(x))
print(bool(x))
print(y)
print(type(y))
print(bool(y))
```

```python
# dict 함수로 딕셔너리 만들기

lux1 = dict(health=400, mana=334, melee=550, armor=10.72)
print(lux1)
print(type(lux1))
```

```python
# zip 함수
lux2 = zip(["health", "mana", "melee", "armor"], [490, 334, 550, 18.72])
print(lux2)
print(type(lux2))

lux2 = dict(zip(["health", "mana", "melee", "armor"], [490, 334, 550, 18.72]))
print(lux2)
print(type(lux2))
```

```python
lux3 = [("health", 490), ("mana", 334), ("melee", 550), ("armor", 18.72)]
print(lux3)
print(type(lux3))
lux3 = dict([("health", 490), ("mana", 334), ("melee", 550), ("armor", 18.72)])
print(lux3)
print(type(lux3))
```

```python
# 딕셔너리 키에 값 할당하기

lux = {
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}
print(lux["health"])
print(lux["armor"])
```

```python
lux["health"] = 2037
print(lux)
```

```python
lux["mana_regen"] = 3.28
print(lux)
```

```python
# 딕셔너리의 키의 개수 구하기

print(lux)
print(len(lux))
```
