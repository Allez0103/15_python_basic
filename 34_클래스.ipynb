```python
# 인스턴스를 만들때 값 받기
# class 클래스이름:
#     def __init__(self, 매개변수1, 매개변수2):
#         self.속성1 = 매개변수1
#         self.속성2 = 매개변수2

class Person:

    def __init__(self, name, age, address):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print(f"{self.hello} 저는 {self.name} 입니다. ")

Maria = Person("Maria", 25, "서울시 서초구 반포동")
Maria.greeting()
print(f"이름: {Maria.name}")
print(f"나이: {Maria.age}")
print(f"주소: {Maria.address}")
```

```python
# 비공개 속성 사용하기
# class 클래스이름:
#     def __init__(self, 매개변수1):
#         self.속성1 = 매개변수1
#         self.__속성2 = 값     # __속성 -> 비공개 (클래스 내에서만 사용)


class Person:

    def __init__(self, name, age, address, wallet):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet  # 비공개속성
    
    def pay(self, amount):
        self.__wallet -= amount
        print(f"남은 금액: {self.__wallet}")
    
Maria = Person("Maria", 20, "서울시 서초구 반포동", 10000)
# Maria.__wallet        클래스 밖에선 가져올 수 없다.
Maria.pay(3000)
```

```python
class Person:

    def __init__(self, name, age, address, wallet):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet  # 비공개속성
    
    def pay(self, amount):
        if amount > self.__wallet:
            print("잔액이 부족합니다.")
            return
        self.__wallet -= amount
        print(f"남은 금액: {self.__wallet}")

Maria = Person("Maria", 20, "서울시 서초구 반포동", 10000)
Maria.pay(3000)
Maria.pay(11000)
```

```python
class Knight:
    def __init__(self, health, mana, armor):
        self.health = health
        self.mana = mana
        self.armor = armor
    def slash(self):
        print("베기!")

x = Knight(health=542.4, mana=210.3, armor=38)
print(x.health, x.mana, x.armor)
x.slash()
```

```python
# 511.68 334.0 298
# 1803.68 1184.0 645

class Annie:
    def __init__(self, health, mana, ability_power):
        self.health = health
        self.mana = mana
        self.ability_power = ability_power

    def tibbers(self):
        damage = (self.ability_power * 0.65) + 400
        print(f"티버: 피해량 {damage}")

health, mana, ability_power = map(float, input().split())

x = Annie(health=health, mana=mana, ability_power=ability_power)
x.tibbers()
```
