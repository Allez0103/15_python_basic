```python
# 인스턴스를 만들때 값 받기
# class 클래스이름:
#     def __init__(self, 매개변수1, 매개변수2):
#         self.속성1 = 매개변수1
#         self.속성2 = 매개변수2

class Person:

    def __init__(self, name, age, address):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print(f"{self.hello} 저는 {self.name} 입니다. ")

Maria = Person("Maria", 25, "서울시 서초구 반포동")
Maria.greeting()
print(f"이름: {Maria.name}")
print(f"나이: {Maria.age}")
print(f"주소: {Maria.address}")
```

```python
# 비공개 속성 사용하기
# class 클래스이름:
#     def __init__(self, 매개변수1):
#         self.속성1 = 매개변수1
#         self.__속성2 = 값     # __속성 -> 비공개 (클래스 내에서만 사용)


class Person:

    def __init__(self, name, age, address, wallet):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet  # 비공개속성
    
    def pay(self, amount):
        self.__wallet -= amount
        print(f"남은 금액: {self.__wallet}")
    
Maria = Person("Maria", 20, "서울시 서초구 반포동", 10000)
# Maria.__wallet        클래스 밖에선 가져올 수 없다.
Maria.pay(3000)
```

```python
class Person:

    def __init__(self, name, age, address, wallet):
        self.hello = "안녕하세요."
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet  # 비공개속성
    
    def pay(self, amount):
        if amount > self.__wallet:
            print("잔액이 부족합니다.")
            return
        self.__wallet -= amount
        print(f"남은 금액: {self.__wallet}")

Maria = Person("Maria", 20, "서울시 서초구 반포동", 10000)
Maria.pay(3000)
Maria.pay(11000)
```

```python
class Knight:
    def __init__(self, health, mana, armor):
        self.health = health
        self.mana = mana
        self.armor = armor
    def slash(self):
        print("베기!")

x = Knight(health=542.4, mana=210.3, armor=38)
print(x.health, x.mana, x.armor)
x.slash()
```

```python
# 511.68 334.0 298
# 1803.68 1184.0 645

class Annie:
    def __init__(self, health, mana, ability_power):
        self.health = health
        self.mana = mana
        self.ability_power = ability_power

    def tibbers(self):
        damage = (self.ability_power * 0.65) + 400
        print(f"티버: 피해량 {damage}")

health, mana, ability_power = map(float, input().split())

x = Annie(health=health, mana=mana, ability_power=ability_power)
x.tibbers()
```

```python
import random
names = [
    "권석현", "김규원", "김성준", "김태훈", "김형민", "문홍식", "박규민", "박성민", "박진영", "박찬현",
    "박희재", "손성욱", "신재민", "오은진", "원영배","이도영", "이수현", "이원석", "이진욱", "조현우", "홍득훈"
]
student_data = {}
for name in names:
    student_data[name] = [random.randint(0, 100) for _ in range(10)]

class student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    def total(self):
        return sum(self.scores)
    def average(self):
        return self.total() / len(self.scores)
    def max_score(self):
        return max(self.scores)
    def min_score(self):
        return min(self.scores)
    def is_pass(self):
        if self.average() > 70:
            return "합격"
        else:
            return "불합격"
    def print_info(self):
        print(f"이름: {self.name}")
        print(f"총점: {self.total()}")
        print(f"평균: {self.average()}")
        print(f"최고점: {self.max_score()}")
        print(f"최저점: {self.min_score()}")
        print(f"{self.is_pass()}")

권석현 = student("권석현", student_data["권석현"])
김규원 = student("김규원", student_data["김규원"])
김성준 = student("김성준", student_data["김성준"])
김태훈 = student("김태훈", student_data["김태훈"])
김형민 = student("김형민", student_data["김형민"])
문홍식 = student("문홍식", student_data["문홍식"])
박규민 = student("박규민", student_data["박규민"])
박성민 = student("박성민", student_data["박성민"])
박진영 = student("박진영", student_data["박진영"])
박찬현 = student("박찬현", student_data["박찬현"])
박희재 = student("박희재", student_data["박희재"])
손성욱 = student("손성욱", student_data["손성욱"])
신재민 = student("신재민", student_data["신재민"])
오은진 = student("오은진", student_data["오은진"])
원영배 = student("원영배", student_data["원영배"])
이도영 = student("이도영", student_data["이도영"])
이수현 = student("이수현", student_data["이수현"])
이원석 = student("이원석", student_data["이원석"])
이진욱 = student("이진욱", student_data["이진욱"])
조현우 = student("조현우", student_data["조현우"])
홍득훈 = student("홍득훈", student_data["홍득훈"])
```
