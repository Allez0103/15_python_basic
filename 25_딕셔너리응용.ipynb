```python
x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x.setdefault('e')
print(x)
x.setdefault('f', 100)
print(x)
```

```python
# 딕셔너리에서 키 값 수정하기

x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x.update(a = 90)
print(x)
x.update(e = 50)
print(x)
x.update(a = 900, f = 60)
print(x)
```

```python
# 딕셔너리 표현식 (컴프리헨션)

# 1. {키 : 밸류 for 키, 밸류 in 딕셔너리}
# 1. dict({키 : 밸류 for 키, 밸류 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']
x = {key : value for key, value in dict.fromkeys(keys).items()}
print(x)
```

```python
keys = ['a', 'b', 'c', 'd']
y = {key: 0 for key in dict.fromkeys(keys).keys()}
print(y)
```

```python
keys = ['a', 'b', 'c', 'd']
values = [10, 20, 30, 40]
x = {key : value for key in keys for value in values}
print(x)
```

```python
keys = ['a', 'b', 'c', 'd']
values = [10, 20, 30, 40]
x = {key : values[keys.index(key)] for key in keys}
print(x)
```

```python
keys = ['a', 'b', 'c', 'd']
values = [10, 20, 30, 40]
x = {keys[i] : values[i] for i in range(len(keys))}
print(x)
```

```python
# 딕셔너리의 할당과 복사

x = {'a' : 0, 'b' : 0, 'c' : 0, 'd' : 0}
y = x
```

```python
print(x is y)
print(id(x))
print(id(y))
```

```python
y['a'] = 99
print(x)
print(y)
```

```python
x = {'a' : 0, 'b' : 0, 'c' : 0, 'd' : 0}
y = x.copy()

print(x is y)
y['a'] = 99
print(x)
print(y)
```

```python
# 중첩 딕셔너리의 할당과 복사

x = {'a' : {"python" : 2.7}, 'b' : {"python" : 3.6}}
y = x.copy()

y['a']['python'] = "2.7.15"

print(x is y)
print(x)
print(y)
```

```python
import copy

x = {'a' : {"python" : 2.7}, 'b' : {"python" : 3.6}}
y = copy.deepcopy(x)

y['a']['python'] = "2.7.15"

print(x)
print(y)
```
