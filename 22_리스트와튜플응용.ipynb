```python
# while 반복문으로 요소 출력하기

a = [38, 21, 53, 62, 19]
i = 0

while i < len(a):
    print(a[i])
    i += 1
```

```python
# 리스트의 가장 작은 수, 가장 큰 수, 합계 구하기

a = [38, 21, 53, 62, 19]

smallest = a[0]
for i in a:
    if i < smallest:
        smallest = i

print(smallest)
# min 함수의 로직
```

```python
a = [38, 21, 53, 62, 19]

largest = a[0]
for i in a:
    if i > largest:
        largest = i

print(largest)
# max 함수의 로직
```

```python
a = [38, 21, 53, 62, 19]

print(min(a))
print(max(a))
```

```python
# 요소의 합계 구하기

a = [10, 10, 10, 10, 10]

total = 0

for i in a:
    total += i

print(total)
# sum 함수의 원리
```

```python
a = [10, 10, 10, 10, 10]

print(sum(a))
```

```python
# 리스트 표현식 사용하기 (컴프리헨션)

# 1. [식 for 변수 in 리스트]
# 2. list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print(a)

b = list(j for j in range(15))
print(b)
```

```python
c = [i + 5 for i in range(10)]
print(c)
```

```python
d = [j * 2 for j in range(10)]
print(d)
```

```python
# 리스트 표현식에서 조건문 사용하기

# 1. [식 for 변수 in 리스트 if 조건식]
# 2. list(식 for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)
```

```python
b = [j + 5 for j in range(10) if j % 2 == 1]
print(b)
```

```python
# for 문과 if 문 여러번 사용하기

# [식 for 변수1 in 리스트1 if 조건식1 
#  for 변수2 in 리스트2 if 조건식2]

a = [
    i * j
    for j in range(2, 10)
    for i in range(1, 10)
]
print(a)
```

```python
# 리스트에 map 사용하기

# list(map(함수, 리스트))
# tuple(map(함수, 튜플))

a = [1.2, 2.5, 3.7, 4.6]
for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```

```python
a = ["4", "2", "3", "1"]
for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```
