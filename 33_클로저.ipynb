```python
# 변수의 사용범위 알아보기

x = 10          # 전역변수
def foo():
    print(x)

foo()
print(x)
```

```python
x = 10
def foo():
    x = 20
    print(x)

foo()
print(x)
```

```python
# global 전역변수

x = 10
print(id(x))
def foo():
    global x        # 전역변수 x를 함수내에서 사용하겠다고 선언
    print(id(x))
    x = 20
    print(id(x))
    print(x)

foo()
print(x)
```

```python
# 함수 안에서 함수 만들기

# def 함수이름1():
#     수행할코드1
#     def 함수이름2():
#         수행할코드2

def print_hello():
    hello = "Hello, World!"
    def print_message():
        print(hello)
    print_message()

print_hello()
```

```python
# 지역변수 변경하기

def A():
    x = 10
    def B():
        x = 20
    B()
    print(x)

A()

```

```python
# nonlocal 지역변수

def A():
    x = 10
    def B():
        nonlocal x
        x = 20
    B()
    print(x)

A()
```

```python
# nonlocal이 변수를 찾는 순서

def A():
    x = 10
    y = 100
    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x += 30
            y += 300
            print("1", x)
            print("2", y)
        print("3", x)
        C()
    B()
    print("4", x)
    print("5", y)
A()
```

```python
# global로 전역변수 사용하기

x = 1
def A():
    x = 10
    def B():
        x = 20
        def C():
            global x
            x += 30
            print("1", x)
        C()
        print("2", x)
    B()
    print("3", x)

A()
print("4", x)
```

```python
# 클로저 사용하기

def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add

c = calc()
print(c(1), c(2), c(3), c(4), c(5))
print(list(map(c, [1, 2, 3, 4, 5])))
```

```python
# lambda로 클로저 만들기

def calc():
    a = 3
    b = 5
    return lambda x: a * x + b

c = calc()
print(c(1), c(2), c(3), c(4), c(5))
print(list(map(c, [1, 2, 3, 4, 5])))
```

```python
# 클로저의 지역변수 변경

def calc():
    a = 3
    b = 5
    total = 0
    def mul_add(x):
        nonlocal total
        total += a * x + b
        print(total)
    return mul_add

c = calc()
c(1)
c(2)
c(3)
```
