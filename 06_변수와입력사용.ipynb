```python
## 변수 여러개 한번에 만들기

x, y, z = 10, 20, 30
print(x)
print(y)
print(z)
```

```python
x = y = z = 10

print(x)
print(y)
print(z)
```

```python
x, y = 10, 20
x, y = y, x

print(x)
print(y)
```

```python
a = input("첫번째 숫자: ")
b = input("두번째 숫자: ")

print(a + b)
```

```python
## 입력값을 정수로 변환

a = int(input("첫번째 숫자: "))
b = int(input("두번째 숫자: "))

print(a + b)
```

```python
## 입력값을 변수 2개에 저장하기

a, b = input("문자열 두개를 입력하세요:").split()
```

```python
print(a)
print(b)
```

```python
# "hello python".split()
a, b = ['hello', 'python']
print(a)
print(b)
```

```python
## 두 숫자의 합 구하기

a, b = input("숫자 2개를 입력하세요:").split()
print(a)
print(b)
print(a + b)
```

```python
a, b = input("숫자 2개를 입력하세요:").split()
a = int(a)
b = int(b)
print(a)
print(b)
print(a + b)
```

```python
## 입력받은 값을 , 를 기준으로 분리하기

a, b = map(int, input("숫자 2개를 입력하세요:").split())

# a, b = map(int, ["10", "20"])

# a, b = 10, 20

print(a + b)
```

```python
a, b = map(int, input("숫자 2개를 입력하세요:").split(","))
print(a + b)
```

```python
result = input("숫자 2개를 입력하세요:").split(",")
# "10", "20".split(",")
print(result)
```
