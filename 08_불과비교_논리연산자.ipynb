```python
## 논리연산자 사용

print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
not True and False or not False
# not 우선, 나머지는 순차적으로
# ((not True) and False) or (not False)
```

```python
# 논리연산자와 비교 연산자 사용

print(10 == 10 and 10 != 5)
# (10 == 10) and (10 != 5)
# 비교연산자가 먼저

print((10 > 5) or (10 < 3))
print(not 10 > 5)
print(not 1 is 1.0)
```

```python
print(not 10)
print(not 0)
print((not 10) > 5)
print((not 10) == 0)
```
