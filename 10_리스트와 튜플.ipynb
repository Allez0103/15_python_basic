```python
# 리스트에 여러가지 자료형 저장하기

person = ["James", 17, 175.3, True]
print(person)
```

```python
# 비어있는 리스트

a = []
b = list()

print(bool(a))
print(bool(b))
```

```python
# range 를 사용하여 리스트만들기

range(10)
```

```python
a = list(range(10))
print(a)
```

```python
b = list(range(0, 10))
print(b)
```

```python
c = list(range(5, 12))
print(c)
```

```python
# range(시작, 끝, 증가폭)
d = list(range(-4, 10, 2))
print(d)
```

```python
e = list(range(10, 10))
print(e)
```

```python
# 증가폭을 음수로 하면
f = list(range(10, 0, -1))
print(f)
```

```python
## 튜플 사용하기

# 튜플 = (값1, 값2, 값3, ...)
# 튜플 = 값1, 값2, 값3, ...     -> 괄호 생략가능

a = (1)
print(type(a))
a = (1, )
print(type(a))
# 하나의 요소를 갖는 튜플을 만들 때는 , 있어야 함 (int와 구분)
```

```python
person = ("James", 17, 175.5, True)
print(type(person))
```

```python
# person = ["James", 17, 175.3, True]
# person = ("James", 17, 175.5, True)
# 리스트의 요소는 수정, 변경이 가능하나 튜플은 안된다
```

```python
# range를 사용하여 튜플 만들기

a = tuple(range(10))
print(a)
```

```python
b = tuple(range(5, 12))
print(b)
```

```python
c = tuple(range(-4, 10, 3))
print(c)
```

```python
# 튜플을 리스트로 만들기, 리스트를 튜플로 만들기

a = [1, 2, 3]
print(a)
a = tuple(a)
print(a)
```

```python
b = (4, 5, 6)
print(b)
b = list(b)
print(b)
```

```python
a = list(range(5, -10, -2))
print(a)
```
