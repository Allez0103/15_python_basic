```python
# 집합 연산 사용

# 합집합
# 세트1 | 세트2
# set.union(세트1, 세트2)

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)
print(set.union(a, b))
print(a.union(b))
```

```python
# 교집합
# 세트1 & 세트2
# set.intersection(a, b)

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a & b)
print(set.intersection(a, b))
print(a.intersection(b))
```

```python
# 차집합
# 세트1 - 세트2
# set.difference(a, b)

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a - b)
print(set.difference(a, b))
print(a.difference(b))

print("=" * 10)

print(b - a)
print(set.difference(b, a))
print(b.difference(a))
```

```python
# 대칭 차집합
# 세트1 ^ 세트2
# set.symmetric_difference(a, b)

# 교집합 제외한 합집합

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a ^ b)
print(set.symmetric_difference(a, b))
print(a.symmetric_difference(b))
print(b.symmetric_difference(a))
```

```python
## 집합 연산 후 할당 연산자 사용하기

# 세트1 |= 세트2
# 세트1.update(세트2)

a = {1, 2, 3, 4}
a |= {5}
print(a)

a.update({6})
print(a)
```

```python
# 세트1 &= 세트2
# 세트1.intersection_update(세트2)

a = {1, 2, 3, 4}
a &= {0, 1, 2, 3, 4}
print(a)

a.intersection_update({0, 1, 2, 4})
print(a)

set.intersection_update(a, {0, 1, 3, 4, 5})
print(a)
```

```python
# 세트1 -= 세트2
# 세트1.difference_update(세트2)

a = {1, 2, 3, 4}
a -= {3, 4, 5, 6}
print(a)

a = {1, 2, 3, 4}
a.difference_update({3})
print(a)

a = {1, 2, 3, 4}
set.difference_update(a, {3, 4})
print(a)
```

```python
# 세트1 ^= 세트2
# 세트1.symmetric_difference_update(세트2)

a = {1, 2, 3, 4}
a ^= {3, 4, 5, 6}
print(a)

a = {1, 2, 3, 4}
a.symmetric_difference_update({3, 4, 5, 6})
print(a)

a = {1, 2, 3, 4}
set.symmetric_difference_update(a, {3, 4, 5, 6})
print(a)
```

```python
a = {1, 2, 3, 4}
a.intersection({3, 4})      # 재할당이 되는건 아님
print(a)

a.intersection_update({3, 4})   # 재할당
print(a)
```

```python
# 부분집합과 상위 집합 확인하기

# 현재세트 <= 다른세트
# 현재세트.issubset(다른세트)

a = {1, 2, 3, 4}
print(a <= {4, 3, 2, 1})
print(a.issubset({2, 3, 4, 5, 1}))
```

```python
# 현재세트 < 다른세트

a = {1, 2, 3, 4}
print(a < {2, 1, 4, 3, 5})
print(a < {2, 1, 4, 3})
```

```python
# 현재세트 >= 다른세트
# 현재세트.issuperset(다른세트)

a = {1, 2, 3, 4}
print(a >= {1, 2, 3, 4})
print(a >= {1, 2, 3})
print(a.issuperset({1, 2, 3, 4}))
```

```python
# 현재세트 > 다른세트

a = {1, 2, 3, 4}
print(a > {1, 2, 3, 4})
print(a > {1, 2, 3})
```

```python
# 세트가 같은지 다른지 확인

a = {2, 3, 4, 1}
print(a == {1, 2, 4, 3})
print(a == {4, 2, 1, 3})
print(a == {4, 2, 1})
```

```python
# 세트간에 겹치는 요소가 있는지 확인

# 현재세트.isdisjoint(다른세트)

a = {1, 2, 3, 4}
print(a.isdisjoint({5, 6, 7, 8}))       # 겹치는 부분이 없으면 True
print(a.isdisjoint({3, 4, 5, 6}))       # 겹치는 부분이 있으면 False
```

```python
# 세트에 요소 추가하기

a = {1, 2, 3, 4}
a.add(5)
print(a)

a.add(4)
print(a)
```

```python
# 세트에서 특정 요소 삭제

a = {1, 2, 3, 4}
a.remove(3)
print(a)

# a.remove(6)       없는 요소를 지우려 할 경우 에러

a.discard(2)
print(a)

a.discard(6)        # discard 를 활용할 경우 에러안남
print(a)
```

```python
# 세트의 할당과 복사 (참조)

a = {1, 2, 3, 4}
b = a
print(a is b)

b.add(5)
print(a)
print(b)
```

```python
a = {1, 2, 3, 4}
b = a.copy()

print(a is b)
print(a == b)

b.add(5)
print(a)
print(b)
```

```python
# 세트 표현식 사용하기 (컴프리헨션)

a = {i for i in 'apple'}
print(a)

b = {j for j in range(5)}
print(b)
```

```python
# 세트 표현식에서 if 조건문 사용하기
# 1. {식 for 변수 in 세트 if 조건식}
# 2. set(식 for 변수 in 세트 if 조건식)

a = {i for i in 'pineapple' if i not in 'apl'}
print(a)
```

```python
a = {i for i in range(1, 100 + 1) if i % 3 == 0}
b = {i for i in range(1, 100 + 1) if i % 5 == 0}
print(a & b)
```

```python
a, b = map(int, input().split())

set_a = {i for i in range(1, a + 1) if a % i == 0}
set_b = {j for j in range(1, b + 1) if b % j == 0}

divisor = set_a & set_b
result = 0
if type(divisor) == set:
    result = sum(divisor)

print(result)
```
