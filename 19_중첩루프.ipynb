```python
# 사각형으로 * 출력하기

for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
    
```

```python
num = int(input())

for i in range(num):
    for j in range(num):
        print("*", end=" ")
    print()
```

```python
# 사각형 모양 바꾸기

for i in range(3):
    for j in range(7):
        print("*", end=" ")
    print()
```

```python
# 계단식으로 별 출력하기

for i in range(5):
    for j in range(5):
        if j <= i:
            print("*", end=" ")
    print()
```

```python
for i in range(5):
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

```python
# 대각선으로 * 출력하기

for i in range(5):
    for j in range(5):
        if j == i:
            print("*", end=' ')
        else:
            print(" ", end=' ')
    print()
```

```python
for i in range(5):
    for j in range(i + 1):
        if j == i:
            print("*", end=' ')
        else:
            print(" ", end=' ')
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j >= i:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
height = int(input())

for i in range(height):
    for j in range(height - i):
        print(" ", end="")
    for k in range(i + 1):
        print("*", end="")
    for l in range(i):
        print("*", end="")
        
    print()
        
```
