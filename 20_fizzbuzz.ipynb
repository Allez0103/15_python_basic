```python
# 1 부터 100 까지 숫자 출력하기

for i in range(1, 100 + 1):
    print(i)
```

```python
# 3의 배수는 fizz 5의 배수는 buzz

for i in range(1, 100 + 1):
    if i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)
```

```python
# 3과 5의 공배수 처리

for i in range(1, 100 + 1):
    if i % 3 == 0 and i % 5 == 0:
        print('fizzbuzz')
    elif i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)
```

```python
# 논리연산자 없이 공배수 처리하기

for i in range(1, 100 + 1):
    if i % 15 ==0:
        print('fizzbuzz')
    elif i % 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('buzz')
    else:
        print(i)
```

```python
# 코드 단축하기

for i in range(1, 101):
    print('fizz' * (i % 3 == 0) + 'buzz' * (i % 5 == 0) or i)
```

```python
for i in range(1, 101):
    if i % 2 == 0 and i % 11 == 0:
        print('fizzbuzz')
    elif i % 2 == 0:
        print('fizz')
    elif i % 11 == 0:
        print('buzz')
    else:
        print(i)
```
