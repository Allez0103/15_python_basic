```python
## for 반복문

# "Hello, World!"를 100번 반복
# print("Hello, World!") 100번 타이핑
# print("Hello, World!\n" * 100)
```

```python
# for와 range 사용하기

# for 변수 in range(횟수):
#     반복할 코드

for _ in range(5):
    print("Hello, World!")
```

```python
# 반복문에서 변수의 변화 알아보기

for i in range(10):
    print("Hello, World!", i, sep=":")

print(i)
```

```python
for i in range(5, 12):
    print("Hello, World!", i, sep="-")
```

```python
for imsi in range(0, 10, 2):
    print("Hello, World!", imsi, sep="_")
```

```python
for a in range(10, 0):
    print("Hello, World!", a, sep="/")
```

```python
print(list(range(10, 0)))
```

```python
for a in range(10, 0, -1):
    print("Hello, World!", a, sep="/")
```

```python
print(list(range(10)))
print(list(reversed(range(10))))
```

```python
for c in reversed(range(10)):
    print("Hello, World!", c, sep="~")
```

```python
# 시퀀스 객체로 반복하기

a = [10, 20, 30, 40, 50]

for imsi in a:
    print(imsi)
```

```python
fruits = ('apple', 'orange', 'grape')

for fruit in fruits:
    print(fruit)
```

```python
for letter in "Python":
    print(letter)
```

```python
x = [49, -17, 25, 102, 8, 62, 21]

for ten in x:
    print(ten * 10, end=" ")
```

```python
dan = int(input())

for i in range(1, 10):
    print(dan, "*", i, "=", dan * i, sep=" ")
```
