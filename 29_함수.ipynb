```python
## 함수 만들기

# def 함수이름():
#     수행할코드

def hello():
    print("Hello, World!")
```

```python
# 함수 호출하기

hello()
```

```python
# 덧셈 함수 만들기

# def 함수이름(매게변수1, 매게변수2):
#     수행할코드

def add(a, b):
    print(a + b)
```

```python
add(10, 20)
```

```python
# 함수 결과 반환하기

def add(a, b):
    return a + b
```

```python
x = add(15, 85)
print(x)
```

```python
# 함수에서 값 여러개 반환하기

def add_sub(a, b):
    return a + b, a - b
```

```python
add_sub(10, 20)
```

```python
a = add_sub(10, 20)
print(a)
```

```python
x, y = add_sub(15, 10)
print(x, y)
```

```python
# 함수의 호출과정 보기

def mul(a, b):
    c = a * b
    return c

def add(a, b):
    c = a + b
    print(c)
    d = mul(a, b)
    print(d)

x, y = 10, 20
add(x, y)
```

```python
x = 10
y = 3

def get_quotient_reminder(a, b):
    return a // b, a % b

quotient, reminder = get_quotient_reminder(x, y)
print(f"몫: {quotient}, 나머지: {reminder}")
```

```python
x, y = map(int, input().split())

def calc(a, b):
    return a+b, a-b, a*b, a/b

a, s, m, d = calc(x, y)
print(f"덧셈: {a}, 뺄셈: {s}, 곱셈: {m}, 나눗셈: {d:.1f}")
```
