```python
price = int(input("가격: "))
coupon = input("쿠폰: ")

if coupon == 'Cash3000':
    print(price - 3000)

if coupon == 'Cash5000':
    print(price - 5000)

if coupon != 'Cash3000' and coupon != 'Cash5000':
    print(price)
```
