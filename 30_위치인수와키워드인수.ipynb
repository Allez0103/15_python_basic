```python
## 위치인수를 사용하는 함수 만들고 호출하기

def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30)
```

```python
# 언패킹 사용하기

x = [10, 20, 30]
print_numbers(*x)
```

```python
y = (10, 20, 30)
print_numbers(*y)
```

```python
print_numbers(*[5, 10, 15])
print_numbers(*(5, 10, 15))
```

```python
# 가변인수 만들기

# def 함수이름(*매개변수):
#     수행할문장

def print_number(*args):
    print(args)
    for arg in args:
        print(arg)
```

```python
print_number(10)
```

```python
print_number(10, 20, 30, 40, 50)
```

```python
print(10, 20, 30, 40)
```

```python
x = [5, 10, 15]
x = print_number(*x)
```

```python
y = [10, 20, 30, 40]
print_number(*y)
```

```python
def print_numberss(a, *args):
    print(a)
    print(args)
```

```python
print_numberss(1)
```

```python
print_numberss(1, 10, 20)
```

```python
print_numberss([1, 2, 3], [4, 5, 6])
```

```python
# 키워드 인수 사용하기

def personal_info(name, age, address):
    print("이름", name)
    print("나이", age)
    print("주소", address)
```

```python
personal_info("홍길동", 30, "서울시 용산구")
```

```python
personal_info(name="홍길동", age=30, address="서울시 용산구")
```

```python
personal_info(age=25, address="대구시 동구", name="김철수")
```

```python
x = {"name" : "홍길동", "age" : 30, "address" : "서울시 용산구"}
personal_info(**x)
```

```python
# 키워드 인수를 사용하는 가변인수 함수 만들기

# def 함수이름(**매개변수):
#     수행할코드

def personal_info(**kwargs):
    print(kwargs)
    for kw, arg in kwargs.items():
        print(kw, ": ", arg, sep="")
```

```python
personal_info(name = "홍길동")
```

```python
personal_info(name="홍길동", age=30, address="서울시 용산구")
```

```python
# 매개변수에 초기값 저장하기

# def 함수이름(매개변수 = 값):
#     수행할코드

def personal_info(name, age, address = "비공개"):
    print("이름", name)
    print("나이", age)
    print("주소", address)
```

```python
# 70 80 85 38 43 30
# 100 18 38 94 93 12 39

input_data = list(map(int, input().split()))

def calcul(*args):
    data_list = []
    for j in range(len(args[0])):
        data_list.append(args[0][j])

    max_sc = max(data_list)
    min_sc = min(data_list)
    total_sc = sum(data_list)
    avg_sc = total_sc/len(data_list)

    v_num = 0
    for i in data_list:
        v_num += (i - avg_sc)**2
    var_sc = v_num / len(data_list)

    return max_sc, min_sc, total_sc, avg_sc, var_sc

max_score, min_score, total_score, avg_score, var_score = calcul(input_data)
print(f"최고점수: {max_score:.2f}, 최저점수: {min_score:.2f}, 총점: {total_score:.2f}, 평균: {avg_score:.2f}, 분산: {var_score:.2f}")
```

```python
# 아메리카노=2, 카페라떼=1
# 바닐라라떼=2, 초코케이크=1, 샌드위치=1
# 아메리카노=1, 샌드위치=2

prices = {
    "아메리카노": 2500,
    "카페라떼": 3500,
    "바닐라라떼": 4000,
    "초코케이크": 5500,
    "샌드위치": 6000
}

def order(**kwargs):
    total_price = 0
    for i in kwargs.keys():
        if i in prices.keys():
            product_price = kwargs[i] * prices[i]
            print(f"{i} {kwargs[i]}개: {product_price}원")
            total_price += product_price
        else:
            print(f"등록되지 않은 상품입니다: {i}")
    print(f"총 결제 금액: {total_price}원")
    print("-" * 20)
    
order(아메리카노=2, 카페라떼=1)
order(바닐라라떼=2, 초코케이크=1, 샌드위치=1)
order(아메리카노=1, 샌드위치=2)
order(아메리카노=1, 허니브레드=2)
```
