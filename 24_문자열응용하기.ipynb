```python
# 문자열 위치 찾기

print("apple pineapple".find('pl'))
print("apple pineapple".find('xy'))     # 없을경우 -1 반환
```

```python
# 문자열 포매팅
# "%s" % "문자열"

print("I am %s" % 'James')
name = 'Maria'
print("I am %s" % name)
```

```python
# "%d" % 숫자

print("I am %d years old" % 20)
```

```python
# "%f" % 소수

print("%f" % 2.3)       # 고정소수점
print("%0.2f" % 2.3)
print("%.2f" % 2.3)
print("%.3f" % 2.3)
```

```python
# 문자열 정렬

"%10s" % "python"
```

```python
"%-10s" % "python"
```

```python
# 문자열 안에 값 여러개 넣기

print("Today is %d %s" % (3, "April"))
print("Today is %5d %10s" % (3, "April"))
```

```python
# format 메서드 사용하기

print("Hello, {0}".format("World!"))
print("Hello, {0}".format(100))
```

```python
print("Hello {0} {2} {1}".format('Python', 'Script', 3.6))
print("Hello {0} {1} {2}".format('Python', 'Script', 3.6))
print("Hello {0} {1} {1}".format('Python', 'Script', 3.6))
print("Hello {0} {1} {1} {2} {2} {0}".format('Python', 'Script', 3.6))
```

```python
print("Hello {} {} {}".format('Python', 'Script', 3.6))
```

```python
print("Hello {language} {version}".format(language = "python", version = 3.6))
```

```python
# 숫자 개수 맞추기

print("%03d" % 1)
print('{0:03d}'.format(35))
print("%f" % 3.6)
print("%014.6f" % 3.6)
print("%08.2f" % 3.6)
```

```python
a = 12345.98765
print(f"{a}")
print(f"{a:.2f}")
print(f"{a:012.2f}")
print(f"{a:<012.2f}")
print(f"{a:<012,.2f}")
```

```python
# 51900;83000;158000;367500;250000;59200;128500;1304000

num_list = list(map(int, input().split(';')))
num_list = sorted(num_list)
num_list.reverse()

for i in range(len(num_list)):
    print(f"{num_list[i]:,}")
```
