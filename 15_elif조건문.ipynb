```python
## elif 사용하기

x = 20

if x == 10:
    print("10 입니다.")
elif x == 20:
    print("20 입니다.")
else:
    print("no")
```

```python
x = 30

if x == 10:
    print("10 입니다.")
elif x == 20:
    print("20 입니다.")
else:
    print("10도 아니고 20도 아닙니다.")
```

```python
# 음료수 자판기 만들기

button = int(input())

if button == 1:
    print('coke')
elif button == 2:
    print('cider')
elif button == 3:
    print('fanta')
else:
    print('제공하지 않는 메뉴')
```

```python
x = int(input())

if 11 <= x <= 20:
    print('11 ~ 20')
elif 21 <= x <= 30:
    print('21 ~ 30')
else:
    print('아무것도 해당하지 않음')
```

```python
money = 9000
age = int(input())

if age >= 7:
    if 7 <= age <= 12:
        print(money - 650)
    elif 13 <= age <= 18:
        print(money - 1050)
    else:
        print(money - 1250)
else:
    print('wrong input')
```
