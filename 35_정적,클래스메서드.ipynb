```python
## 클래스 속성 사용하기

# class 클래스이름:
#     속성 = 값

class Person:
    bag = []

    def put_bag(self,stuff):
        self.bag.append(stuff)

james = Person()
james.put_bag('책')
print(james.bag)

Maria = Person()
Maria.put_bag('열쇠')

print(james.bag)
print(Maria.bag)

# 인스턴스를 공유한다
```

```python
# 인스턴스 속성 사용하기

class Person:
    # bag = []
    def __init__(self):
        self.bag = []

    def put_bag(self,stuff):
        self.bag.append(stuff)

james = Person()
james.put_bag('책')
print(james.bag)

Maria = Person()
Maria.put_bag('열쇠')

print(james.bag)
print(Maria.bag)
```

```python
# 비공개 클래스 속성

# class 클래스이름:
#     __속성 = 값

class Knight:
    __item_limit = 10

    def print_item_limit(self):
        print(Knight.__item_limit)

x = Knight()
x.print_item_limit()

# print(Knight.__item_limit())
```

```python
# 정적 메서드 사용하기

# class 클래스이름:
#     @staticmethod
#     def 함수이름(매개변수):
#         수행할코드

class calc:
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def mul(a, b):
        return a * b
    
print(calc.add(5, 6))
print(calc.mul(5, 6))
```

```python
# 클래스 메서드 사용하기

# class 클래스이름:
#     @classmethod
#     def 메서드이름(cls, 매개변수1, 매개변수2):
#         수행할코드

class Person:

    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def print_count(cls):
        print(f"{cls.count} 명 생성되었습니다.")
    
james = Person()
maria = Person()

Person.print_count()
```

```python
class Date:
    @staticmethod
    def is_date_valid(date_string):
        year, month, day = map(int, date_string.split("-"))
        return month <= 12 and day <= 31

if Date.is_date_valid('2000-10-31'):
    print('올바른 날짜 형식')
else:
    print('잘못된 날짜형식')
```
