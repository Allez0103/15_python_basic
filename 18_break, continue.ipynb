```python
## break로 반복문 끝내기

i = 0

while True:
    print(i)
    i += 1
    if i == 10:
        break
```

```python
# for에서 break로 반복문 끝내기

for i in range(10000):
    print(i)
    if i == 10:
        break
```

```python
## continue로 코드실행 건너뛰기

for i in range(20):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# while 반복문에서 continue로 코드 실행 건너뛰기

i = 0

while i < 20:
    i += 1
    if i % 2 == 0:
        continue
    print(i)
```

```python
# 입력한 횟수대로 반복하기

count = int(input("number: "))

i = 0

while True:
    i += 1
    print(i)
    if i == count:
        break
```

```python
# 입력한 숫자까지 홀수 출력하기

count = int(input("number: "))

for i in range(count + 1):
    if i % 2 == 0:
        continue
    print(i)
```

```python
i = 0
while True:
    if i % 10 == 3:
        print(i, end=" ")
    if i > 73:
        break
    i += 1
```

```python
start_num, end_num = map(int, input().split())

i = start_num

while True:
    if i % 10 == 3:
        i += 1
        continue
    if i > end_num:
        break
    print(i, end=" ")
    i += 1
```
