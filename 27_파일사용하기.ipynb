```python
## 파일에 문자열 쓰기

# 파일 객체 = open(파일이름, 파일모드)
# 파일객체.write(문자열)
# 파일객체.close()

file = open("./file_test/hello.txt", "w")
file.write("Hello, World!")
file.close()
```

```python
# 파일에서 문자열 읽기

# 변수 = 파일객체.read()

file = open("./file_test/hello.txt", "r")
s = file.read()
print(s)
file.close()
```

```python
# 자동으로 파일객체 닫기

# with open("파일이름", "파일모드") as 파일객체:
#     코드

with open("./file_test/hello.txt", "r") as file:
    s = file.read()
    print(s)
```

```python
# 반복문으로 문자열 여러줄 파일에 쓰고 읽기

with open("./file_test/hello2.txt", "w") as file:
    for i in range(3):
        file.write(f"Hello, World! {i}\n")
```

```python
# 리스트에 들어어있는 문자열 파일에 쓰기

lines = ["안녕하세요\n", "파이썬\n", "코딩도장입니다.\n"]

with open("./file_test/hello3.txt", "w", encoding="utf-8") as file:
    file.writelines(lines)
```

```python
# 파일의 내용을 한줄씩 리스트로 가져오기

# 변수 = 파일객체.readlines()

with open("file_test/hello3.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
    print(lines)
```

```python
# 파일의 내용을 한줄씩 읽기

# 변수 = 파일객체.readline()

with open("file_test/hello3.txt", "r", encoding="utf-8") as file:
    line = None
    while line != "":
        line = file.readline()
        print(line.strip('\n'))
```

```python
# for 반복문으로 파일의 내용을 줄 단위로 읽기

with open("file_test/hello3.txt", "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip("\n"))
```

```python
# 파이썬 객체를 파일에 저장하고 가져오기

import pickle

name = "James"
age = 17
address = "서울시 서초구 반포동"
scores = {
    'korean' : 90,
    'english' : 95,
    'mathmatics' : 85,
    'science' : 82
}

with open("file_test/james.p", "wb") as file:   # wb -> 바이너리 쓰기
    pickle.dump(name, file)
    pickle.dump(age, file)
    pickle.dump(address, file)
    pickle.dump(scores, file)
```

```python
import pickle

with open("file_test/james.p", "rb") as file:
    name = pickle.load(file)
    age = pickle.load(file)
    address = pickle.load(file)
    scores = pickle.load(file)

print(name)
print(age)
print(address)
print(scores)
```
