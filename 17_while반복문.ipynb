```python
## while문 사용하기

# while 조건문:
#     반복할 코드

i = 0

while i < 10:
    print("Hello, World!")
    i += 1
```

```python
# 초기값 1부터 시작하기

i = 1

while i <= 10:
    print("Hello, World!")
    i += 1
```

```python
# 초기값을 감소시키기

i = 10

while i > 0:
    print("Hello, World!")
    i -= 1
```

```python
# 입력한 횟수대로 반복하기

count = int(input("반복할 횟수를 입력하세요: "))

i = 0
while i < count:
    print("Hello, World!", i + 1, sep=":")
    i += 1
```
