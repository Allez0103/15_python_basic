```python
## 문자열안에 따옴표 포함하기

s = "python isn't difficult"
print(s)
```

```python
s = 'He said "Python is easy"'
print(s)
```

```python
s = 'python isn\'t difficult'
print(s)
s = "He said \"Python is easy\""
print(s)
```
