```python
## 사칙연산

print(1 + 1)
print(1 - 1)
print(2 * 2)
print(4 / 2)
# 공백을 넣어주는 이유는 붙였을 때 전체를 변수로 받아들일 가능성이 있음
```

```python
## 나눗셈 후 소숫점 이하를 버리는 몫 연산자 (//)

print(5 / 2)
print(5 // 2)
```

```python
## 나눗셈 후 나머지를 구하는 연산자 (%)

print(5 % 2)
```

```python
## 거듭제곱 연산자 (**)

print(2 ** 10)

# 제곱근을 할 때

print(4 ** 0.5)
print(4 ** (1/2))
print(16 ** (1/4))
```

```python
## 값을 정수로 만들기

print(int(3.14))
print(4 / 2)
print(int(4 / 2))
```

```python
## 객체의 자료형 알아내기 type

print(type(10))
type(10)
```

```python
## 2진수

print(0b110)    #binary
# 1 * 2 ** 2 + 1 * 2 ** 1 + 0 * 2 ** 0
print(0b111111) 
```

```python
## 8진수

print(0o110)    #octa
# 1 * 8 ** 2 + 1 * 8 ** 1 + 0 * 8 ** 0
```

```python
## 16진수

print(0xAF)
# A = 10 B = 11 C = 12 D = 13 E = 14 F = 15
# 10 * 16 ** 1 + 15 * 16 ** 0
```

```python
## 실수 계산하기

print(3.5 + 2.1)
```

```python
print(4.2 + 5)
```

```python
## 값을 실수로 만들기

print(float(5))
print(float())      # 기본값은 0
print(float("5.3")) # 문자열
# print(float("hi"))
```

```python
## 괄호 사용하기

print(2 + 2 * 2)
print((2 + 2) * 2)
```
