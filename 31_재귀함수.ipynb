```python
## 재귀호출 사용하기

# 함수 자신을 호출

def hello():
    print("hello")
    hello()
```

```python
hello()     # 무한루프 에러
```

```python
## 재귀호출의 종료조건 만들기

def hello(count = 1):
    if count == 0:
        return
    print("Hello, World!", count)
    count -= 1
    hello(count)
```

```python
hello(5)
```

```python
hello()
```

```python
hello(15)
```

```python
# 재귀 호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
def is_palindrome(word):
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```

```python
def fiv(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fiv(n - 1) + fiv(n - 2)
```

```python
print(fiv(10))
print(fiv(20))
```
